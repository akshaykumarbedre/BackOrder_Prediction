```python
columns = ["in_transit_qty"]
n_std = 3

# For each column, remove rows that are more than n_std standard deviations away from the mean
for col in columns:
    mean = df[col].mean()
    std = df[col].std()
    df = df[(df[col] >= mean - n_std * std) & (df[col] <= mean + n_std * std)]
```

```python
li=df.dtypes[(df.dtypes=='int64') | (df.dtypes=='float64')].index
df[li]
```

```python
df['went_on_backorder'].value_counts()
#it immbalace data 
```

```python
from sklearn.utils import resample
majer_class=df[df['went_on_backorder']=="No"]
minar_class=df[df['went_on_backorder']=="Yes"]
resample_data=resample(majer_class,n_samples=len(minar_class),random_state=1)
df=pd.concat([resample_data,minar_class])
```

```python
df['went_on_backorder'].value_counts()
#it immbalace data 
```

```python
ch=df.dtypes[(df.dtypes!='int64') & (df.dtypes!='float64')].index
df[ch]
```

```python
for i in ch:
    df[i]=df[i].str.replace("Yes","1")
    df[i]=df[i].str.replace("No","0")
    df[i]=df[i].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
model1=RandomForestClassifier()
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
